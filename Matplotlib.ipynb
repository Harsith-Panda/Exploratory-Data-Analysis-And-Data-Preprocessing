# Matplotlib

## Introduction:

- This is like mother of all visulization lbraries in python. Just like hoe pandas is built over numpy, visualization libraries like seaborn and etc are built on top of matplotlib.

- Analysis based on a single variable (single column) is called `uni-variate analysis` and analysis based on multiple variables (multiple columns) is called `multi-variate analysis`. If two columns or two variable it is `bivariate analysis`.

## Types Of Data

- Numerical Data (Numbers) (How many litres of water did you drink?, Age?, etc)
    - Discrete Data (Finite, countable, whole numbers) (How many ig posts?)(Answer will always be 1, 2 or some whole numbers)
    - Continous Data (Continous data, decimals, fractions) (How many litres of water?) (Answer could take up any value in interval 1.75, 3.25, 8) 

- Categorical Data (Categories / Text / Labels)
    - Nominal (Does not follow any order) (Phone brand)
    - Ordinal (Has a proper ordering hierarchy) (rating)

## 2D-Line Plot
- Used for bivariate analysis
- Used when analysis is made based on two variables (one in x and one in y)
- `categorical -> numerical` or `numerical -> numerical`
- Use case - Time series data (Time series data is a kind of data where you measure one axis or quantity with respect to time) (Sales in each month)

```python
price = [5000, 6000, 3000, 2000, 100000]
year = [2010, 2011, 2012, 2013, 2014]

plt.plot(year, price) # plt.plot(x-axis list, y-axis list)
# x-axis: Mostly contains actegorical column
# y-axis: Mostly contains numerical columns
```

```python
# Using an actual pandas dataframe
batsman = pd.read_csv('Practice_Datasets/mpl-1/sharma-kohli.csv')
batsman.head()
# This datasetcontains IPL Record how much run the player has made in that particular season
```

```python
# How much run that particular player (Virat) has made in each IPL season
plt.plot(batsman['index'],batsman['V Kohli']) # Pass in series of specified axis
```

```python
# Plotting multiple graphs together
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

# Adding titles - The master title says what the  graph is about
plt.title("Rohith Sharma Vs Kohli")

# Adding labels - The label / description for each axis x & y
plt.xlabel("IPL Season") # corresponds to x-axis label
plt.ylabel("Runs Scored") # corresponds to y-axis label

# You can add colours as well to the graph
# Colors can be hexcode as well
# you can modify linestyle as well
# You can apply linewidth and marker size as well
plt.plot(batsman['index'],batsman['V Kohli'], color='red',marker ='o', markersize = 7)
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', marker='o')

# Label parameter used to say what line is used for what
plt.plot(batsman['index'],batsman['V Kohli'], color='red',marker ='o', markersize = 7, label = "Kholi")
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', marker='o', label="Rohith")

# Used to show line details and description 
plt.legend()
# plt.legend(loc="lower left") -> Used to change location og legend

plt.grid() # Activates the grid line

plt.show() 
# Just because this a console the codes are being executed here and printed. If in a code editor it would not
# plt.show() is the main function that prints the grph on the console ot screen
```

## Scatter Plots

- Just like 2D plots, this is also used for bivariate analysis.
- Study two columns together.
- Actual use is to use this for `numerical & numerical` columns. Compare numerical and numerical columns.
- Use case: Finding correlation

### plt.plot() vs plt.scatter() - for scatter plots

- plt.plot(x-axis, y-axis, 'o') is used to make scatter plot from given dataset
- plt.plot() is faster than plt.scatter(). When you have lrger dataset use plt.plot()
- When you want to modify colours styles make better plots use plt.scatter()

```python
batsmans = pd.read_csv('Practice_Datasets/mpl-1/batter.csv')
batsmans.head()

batsmans = batsmans.head(50)
batsmans.head()
```

```python
# Function for scatter plot

plt.scatter(batsmans['avg'], batsmans['strike_rate'], marker = '+') # plt.scatter(x-axis, y-axis)

# Here the scatter plot is between average runs vs strike rate
# strike rate says how hard hitter he is, avg says how reliable he is like (like average run he hits)

# More strike rate says how hard hitting he is, avg says how reliable or garunteed run he take

# Batsman at top right -> avg is also more, strike rate is also more (reliable and desructive at same time
# Batsman at right bottom -> Very reliable more avg, but less desructive
# Batsman at top left -> Very desructive, but poor average (less reliable)
# Batsman at bottom right -> Kinda use less neither hard hitting nor reliable

plt.title("AVG and SR Analysis")

plt.xlabel("Average")
plt.ylabel("SR")
```

## Bar Chart

- x-axis has categorical column and y-axis has some aggrgate opeartion.
- Mostly used used with categorical and numerical columns. in which numerical column is mostly some aggragate function
- Use case: Perform and show aggregate function of specific groups

```python
houses = ["Amritamayi", "Chinmayi", "Jyotirmayi", "Anadamayi"]
no_stds = [250, 120, 240, 430]

plt.bar(houses, no_stds) # plt.bar(x-axis, y-axis) by default vertical bar graph
plt.grid() 
```

```python
# Horizontal bar grap
plt.barh(houses, no_stds)
```

```python
mbar = pd.read_csv('Practice_Datasets/mpl-1/batsman_season.csv')
mbar.head()
```

```python
# Stacked Bar Graph

# In stacked bar graph you use it just like bormal bar graph only that for each you have a extra paramter called bottom
# This bottom parameter says how many places or values are below the current value
# bottom = df['field i+1'] + df['field i+2'] + df['field i+3'] + ... + df['field n']
# Last will have no bottom parameter

plt.bar(mbar['batsman'], mbar['2015'], bottom = mbar['2016'] + mbar['2017'], label = "2015")
plt.bar(mbar['batsman'], mbar['2016'], bottom = mbar['2017'], label = "2016")
plt.bar(mbar['batsman'], mbar['2017'], label = "2017")

plt.legend()
```

## Histogram

- Used for univariate analysis on numerical column
- Bar graph is for categorical column, histogram is for numerical column. It is basically for like performing aggregate of rgroup function and analysis on numerical columns (example finding frequency, counts and etc)
- Let's say you have marks of all students for a particular subject in a particular exam. You have to find how many student have scored between some specified range, at this situation you can use histogram.
- Breaks / splits numerical columns into categories and helps to find aggregation or frequency.
- Use case: Frequency count

```python
vk = pd.read_csv('Practice_Datasets/mpl-1/vk.csv')
vk.head()
```

```python
plt.hist(vk['batsman_runs'], bins = np.linspace(0, vk['batsman_runs'].max() + 7, 10))
```

```python
# Handling bin values

# There could be cases where few ranges of bin having more values dimenishing other ranges
# Hence we could use logarithmic scale for bins
# Linear scale is where interval is like (0-10, 10-20, 20-30) distance between data points are range are same
# Logarithmic Scale uses multiplication x10, x100, x 1000 like that, different distance between data ranges.
# This basically converts y to logarithmic scale

npa = np.load('Practice_Datasets/mpl-1/big-array.npy')
npa[0]

#plt.hist(npa) # This dimenishes few values due to issue in data range

plt.hist(npa,rwidth=0.8, log = True) # Logarithmic scale solves this issue
```

## Pie Chart

- Used for univariate and bivariate analysis.
- It is mostly used to with categorical column. It says out of hundred percent how much contribution does the specific category have.
- Categorical vs Numerical
- Find contribution on standard scale (100 %)

```python
plt.pie(no_stds)

plt.pie(no_stds, labels = houses)
```

```python
pied = pd.read_csv('Practice_Datasets/mpl-1/gayle-175.csv')
pied.head()
```

```python
plt.pie(pied['batsman_runs'], labels = pied['batsman'])

# In the above plot percentage is not shown, we use autopct for that
# autopct='%0.1f%%' 0.1f says show decimal upto 1, 0.2f says show decimal upto 2
plt.pie(pied['batsman_runs'], labels = pied['batsman'], autopct='%0.1f%%')
plt.pie(pied['batsman_runs'], labels = pied['batsman'], autopct='%0.1f%%', colors = ['red','white', 'yellow', 'orange', 'green', 'black'])

# Colors explode everything is applied on the order of data present in dataset
```

```python
# Explode basically pops out that specific entry from the pie graph kind of like pulled out
plt.pie(pied['batsman_runs'], labels = pied['batsman'], autopct='%0.1f%%', explode = [0.2,0,0,0,0,0])

# Shadow is just a piece of styling customization
plt.pie(pied['batsman_runs'], labels = pied['batsman'], autopct='%0.1f%%', explode = [0.2,0,0,0,0,0], shadow=True)
```

## Styling your graph

- `plt.style.available` -> Returns a list which contains names of all possible styles
- `plt.style.use(style_string)` -> Used to apply the specified styles on all plots (graphs)

## Saving and Exporting graph as a png (image)

- `plt.savefig(filename)` -> Used export the graph on this name, like save as this image

```python
plt.style.available
plt.style.use('Solarize_Light2')

plt.pie(pied['batsman_runs'], labels = pied['batsman'], autopct='%0.1f%%', explode = [0.2,0,0,0,0,0], shadow=True)
plt.savefig('saved_graph_test.png')
```

## Advanced Matplotlib (Beyond 2D Graphs)

### Basic plt.figure() and cutomization
- plt.figure(figsize=(10,50)) # plt.figure() -> Used to create a blank canvas, graph followed by this will be applied with cutom setting passed
- figsize(width, height) -> Takes in width and height of canvas

### Annotations
- `df.sample(size, random_state=int)` -> size is the size sample size, random_state will be an integer
- If you use the same integer, it gives the same randomly selected sample
- randomly picked rows of given size
- `replace=True/False` -> This parameter says if duplicate rows are allowed in sample or not
- `random_state = 42` -> Industry standard, makes sure if state is same no matter how many times you generate sample of size 'x' it would be consistent

- `x.title(x_point, y_point, 'title_string')`

```python
df = pd.read_csv('Practice_Datasets/mpl-1/batter.csv')
df.head()

sample_df = df.sample(30, random_state=42)
sample_df.head()
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

# Two ways to use annotations

# 1. Using Itertuples

for r in sample_df.itertuples():
    plt.text(r.avg, r.strike_rate, r.batter)

# 2. Using traditional values
for index in range(sample_df.shape[0]):
   plt.text(sample_df['avg'].values[index],sample_df['strike_rate'].values[index], sample_df['batter'].values[index])
```
