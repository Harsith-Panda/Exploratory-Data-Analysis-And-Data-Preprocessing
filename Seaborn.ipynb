# Seaborn

## Introduction

- Seaborn is basically built upon matplotlib, it provides a layer of abstraction which makes plotting easier.
- Better aesthetics, easier to use, lesser number of codes, More graphs included

### Figure & Axis

- Figure is the container on which graph is present. (fig denoted in most places)
- The actual plot or the graph is said to be axis. (ax denoted in most places)

### Categories
#### 1. Relational
1. Lineplot
2. Scatter plot

#### 2. Distributional
1. Histogram
2. KDE Plot

#### 3. Categorical
1. Bar Graph
2. Pie Chart
