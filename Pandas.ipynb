# Pandas

## Introduction
- It is a library built upon numpy, while numpy is used handle n-dimensional arrray. Pandas speciality or use case is to handle handle tabular form of data like (csv, excel and etc)
- It can handle, manipulate many number of structured datatypes. Html, Sql, CSV, excel and more like that.
- It can't hanle `.mp3`,`.mp4` or other audio formats directly. There are other libraries that do that and convert them back to an audio numpy array
- Pandas gives us to main objects for us to manipulate data
    1. Series -> 1D array (labelled data)
    2. DataFrame -> 2d tabular structure (mainly used for manipulating tabular structure)
- Fundamentally dataframe is a form of 2d structure with rows and columns but a row entry (cell) can have complex dt of ndimension numpy array as well. But fundamentally treated as 2d array (table)
- `Interview definition` -> Pandas is built on NumPy and is designed for structured, tabular data. Its primary objects are Series (1D labeled arrays) and DataFrames (2D labeled tables). While it can ingest data from formats like CSV, Excel, SQL, and HTML, it does not process media formats like MP3 or MP4 directly. DataFrames are fundamentally 2D, though individual cells may contain complex Python objects.
- One column or One row from a table / dataframe is still a series

## Series
- It is basically an object or datastructure provided to us by pandas to create 1d Structured labeled array
- Like a column in excel with row labels
- Having row labels is optional. By default index is used as row labels. It is represented in form of two columns first one in left being the row labels and the ones in the right are corresponding values.
- When you are trying to have custom labels, the number of labels must be equal to the total number of elements
- `pd.Series([1,2,3,4])` or `pd.Series([1,2,3,4], index=[a,b,c,d])`
- Despite having custom labels you can use internal indexing basically indexes to access elements as well
- Having custom column names is called label-based indexing, integer-location based indexing refers to the actual indexes of series
- `iloc` -> Accepts only integer-loaction based indexing
- `loc` -> Accepts both integer based and label based indexing
- Series also as head function like dataframe

## Dataframe
- This is an other object or DS provided to use by pandas
- This is basically of 2d form, but internal cells can have more complex nested structure (dimension)
- Dataframe is basically like a table with rows and columns.
- `iloc` in a data frame includes both rows and columns [rows, columns] for more dimensions same as numpy. Right to left -> top most to inner most

````python
import pandas as pd

# Series

ser1 = pd.Series([1,2,3,4]) # Default labels: Index
ser2 = pd.Series([1,2,3,4,5], index=["a","b","c","d","e"]) # Custom Labels

# You can still access them like standard list
#ser2[0] # Works but gives a warning
ser2["a"] # Works

# Location based indexing
ser1.loc[0]
ser2.loc["a"]

# Integer_Location based indexing
# ser2.iloc["a"] won't work accepts only integer
ser1.iloc[0]

# DataFrame

df1 = pd.DataFrame({"name": ["Harsith","Sathiya","Siva"], "age": [19,34,35]})
df1 = pd.DataFrame({"name": ["Harsith","Sathiya","Siva"], "age": [19,34,35]}, index = ["a","b","c"]) # Custom indexing for rows

df1.loc["a"]
````

# Dropping duplicate values / rows

- We use `drop_duplicates(subset=[])` for dropping duplicate values. It keeps the first time a value occuring in a column and removes remaining row that are repeating.
- subset recieves all the columns on which we want to check. If only one column keeps first unique values and removes remaining repating rows. If two or more performs and operation kind of. When both condition matched many times keeps the first and removes alll remaining repeating values.

````python
# Dropping based on sigle column

df.drop_duplicates(subset=['city']).shape # (31,18) Removed all duplicates

df.drop_duplicates(subset=['city','winner']).shape # (198, 18)

# By default all returns the first unique value
# In cse if we want last repeating value we use (subset=[], keep='last')

df.drop_duplicates(subset=['city', 'winner'], keep='last')
````

## Grouping based on column entries

- It is more intutive to apply group by to categorical columns with repeated values. For example in a dataset of shops sales, sectors of multiple shops could be same hence you can group by sectors
- It is used to perform analysis over groups
- You can apply aggregate function directly to this object and it will give you the corresponding result values. Like mean, min, max etc. This will be applied to each group

````python
# Gruping by a specific column df.groupby('column_name') Group all repeating columns into one

df.groupby('season') # It returns a group by object. Can't access it diretly

groupped = df.groupby('season')
groupped

# Groupping on multiple columns 
# df.groupby(['cn1', 'cn2'])

# Find how many group are present
len(groupped)

# Fing number of entries in each group
groupped.size() # This returns a series with label being category and values being number of data in each category

# First & last entry in group
groupped.first() # Returns a df of all first occuring entries in each group
groupped.last() # Returns df of all last entries in each category

groupped.groups
# This will return a dictionary, where every key represents an entry in the column or one group, value has an array of indexes of entries belonging to that group
df.loc[59]

# Fetch all datas in that specific group
groupped.get_group(2008) # Returns a dataframe of entries specific to that group
df.head(1)
````

## Merge Function

- It is used to cross reference or combine datasets. It is more like joins in SQL.
- Condition is that there must be somekind of relation between two. Like one refering to another
- For example in ours we have `match_id` in delivery and `id` in actual dataframe
- Require a column to connect two datasets, name doesn't matter but must have a relation
- `df_to_merge_with.merge(df_to_merge, left_on='ref_with', right_on='ref_to_merge')`
- By default it does inner join. But by adding `how` parameter you can specify what kind of join you want

````python
delivery.head(1)

# Merge matches with each delivery
# df.merge(what, left_on='basis_of_main', right_on='basis_of_merging')
delivery.merge(df, left_on='match_id', right_on='id') # Inner Join
delivery.merge(df, left_on='match_id', right_on='id', how="left") # Left Join

# Total score made by each batsman in each season
merged = delivery.merge(df, left_on='match_id', right_on='id')
merged.groupby(['batsman', 'season'])['total_runs'].sum()
````

## Handling Missing Values

- When it comes to handling missing values, you have two options.
    1. Remove missing values (dropna())
    2. Filling missing values (fillna())
- In this we will be focusing on how o do it. Rather than focusing on when to do what.

### .dropna()

- It by default iterates row by row, if at any row at any column it finds a missing value it deletes the entire row. Even for a single non-value column removes whole row.

### .fillna()

- It is used to fill null values / missing values.
- fillna() is more appropriate / intutive to use it in a column. Like single column

````python
df.dropna()  # Deletes rows with any missing value at any position

df.dropna(axis = 1) # Deletes the column entirely

df.dropna(subset = ["city"]) # It basically says I don't care about other columns, if any row is empty for this specific column remove that row entirely
# Using multiple columns in subsets acts like or. If this empty or that empty. anything empty remove the row

df.fillna(0) # Basically it fills in with whatever value in bracket to whereever the value is missing

# df['cn1'].fillna(value)

df['city'].fillna('Chennai') # Fills in all places missing values 

df['city'].fillna(method='ffill') # If value missing takes the previous cell value and fills in
df['city'].fillna(method='bfill') # If value missing takes in next cell value and fills in
````

## Map()
- A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values.
- Take one set of values, and convert it to an other set of values.
- we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later
- Map is just like in javascript and java. For every value in the series, a particular function is applied and the value is replaced.
- Returns a new series, for every value apply a specific function and put / replace that value
- `lambda parameter: expression_with_that_parameter`
- The following is a function
```python
def subtract_mean(p):
    return p - review_points_mean
```
- Same function expressed in lambda
- lambda p: p - review_points_mean

## .apply()

- This is used to achieve the same behaviour but instead of a single column or series you can do it over an entire dataframe
```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns') # Reviews here is a dataframe and that entire df is modified. Each time each row passed
```
- Perform the same thing but row wise transformation, use `axis='index'`
